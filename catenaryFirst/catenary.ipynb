## Introduction

The theoretical solution of the catenary is the function $y = a \cosh(\frac{x - x_0}{a}) + y_0$ with $x_0 = 0.5$ the lowest point. The parameter $a$ is the length $L$ of the catenary between $-w/2$ and $w/2$, i.e. $L = 2a \sinh(\frac{w}{2a})$. With $w = 1$ and $L = 2$ we have to find $a$ as the zero of the function $f(x) = 1 - a \sinh(1/2a)$.

```r
f <- function(x) { 1 - x*sinh(1/(2*x)) }
a <- uniroot(f, c(0.001, 1), tol=1e-15)$root
```

To fix the catenary at the end points $(0,1)$ and $(1,1)$ the vertical shift is calculated as

```r
x0 <- 0.5; y0 <- 1 - a*cosh(x0/a)
fnCatenary <- function(x) { a*cosh((x-x0)/a) + y0 }
xs <- seq(0, 1, length.out=51); ys <- fnCatenary(xs)
```

## Augmented Lagrangian

```r
N <- 51         # no. of points
L <- 2          # total length of chain
h <- L / (N-1)  # maximal length of each chain link
```

```r
fnobj <- function(p) {
    sum(p[(N+1):(2*N)])  # sum(y)
}

grobj <- function(p) {
    c(rep(0, N), rep(1, N))
}
```

```r
heq <- function(p) {
    c(p[1], p[N]-1, p[N+1]-1, p[2*N]-1)
}

heq.jac <- function(x) pracma::jacobian(heq, x)
```

```r
hin <- function(p) {
    x <- p[1:N]; y <- p[(N+1):(2*N)]
    h^2 - diff(x)^2 - diff(y)^2
}

hin.jac <- function(x) pracma::jacobian(hin, x)
```

```r
x0 <- seq(0, 1, length.out=N)
p0 <- c(x0, rep(0, N))
```

```r
system.time(
sol1 <- alabama::auglag(p0, fnobj, grobj, hin=hin, heq=heq,
                        control.outer=list(trace=FALSE))
)
x1 <- sol1$par[1:N]; y1 <- sol1$par[(N+1):(2*N)]
```

```r
par(mar=c(2.1,2.1,2.1,1.1))
plot(c(0, 1), c(0, 1), type='n',
     xlab="x", ylab="y", main="Fig. 1: auglag()"); grid()
points(c(0, 1), c(1, 1))
lines(x1, y1, col="blue", lwd=2)

lines(xs, ys, col="red")
```

```r
rmse <- function(y1, y2) { sqrt(sum((y1 - y2)^2)/length(y1)) }
y2 <- fnCatenary(x1)
rmse(y1, y2); abs(y1 - y2); 
```
